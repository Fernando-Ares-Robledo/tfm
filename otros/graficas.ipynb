```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-darkgrid')
light_blue = "mediumturquoise"#"deepskyblue"#'teal'#'#74EDFF'


sizes_gb = np.array([1e3, 1e6, 1e7, 1e8, 1e9, 2e9, 5e9, 1e10])
times_sec = np.array([0.5, 0.5, 0.5, 2, 28, 42, 191, 452])
errors_sec = np.array([0, 0, 0, 1, 5, 20, 40, 120])


plt.figure(figsize=(10, 6))
plt.plot(sizes_gb, times_sec, 'o-', color=light_blue)

plt.fill_between(sizes_gb, times_sec - errors_sec, times_sec + errors_sec, color=light_blue, alpha=0.3)
plt.errorbar(sizes_gb, times_sec, yerr=errors_sec, fmt='o-', capsize=5, capthick=2, color=light_blue)
plt.xlabel('Tamaño del Archivo (bytes)')
plt.ylabel('Tiempo de Backup (segundos)')
plt.title('Velocidades de Backup para Diferentes Tamaños de Archivo')
plt.xscale('log')

plt.grid(True, which="both", ls="--")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd



# Cargar los datos de uso de CPU
cliente_baseline_cpu = pd.read_csv('cliente_baseline_cpu_usage.txt', delim_whitespace=True, skiprows=3, usecols=[2, 3, 4, 5, 6, 7, 8], names=['CPU', '%user', '%nice', '%system', '%iowait', '%steal', '%idle'])
cliente_backup_cpu = pd.read_csv('cliente_backup_cpu_usage.txt', delim_whitespace=True, skiprows=3, usecols=[2, 3, 4, 5, 6, 7, 8], names=['CPU', '%user', '%nice', '%system', '%iowait', '%steal', '%idle'])

server_baseline_cpu = pd.read_csv('server_baseline_cpu_usage.txt', delim_whitespace=True, skiprows=3, usecols=[2, 3, 4, 5, 6, 7, 8], names=['CPU', '%user', '%nice', '%system', '%iowait', '%steal', '%idle'])
server_backup_cpu = pd.read_csv('server_backup_cpu_usage.txt', delim_whitespace=True, skiprows=3, usecols=[2, 3, 4, 5, 6, 7, 8], names=['CPU', '%user', '%nice', '%system', '%iowait', '%steal', '%idle'])

# Cargar los datos de uso de Memoria
cliente_baseline_mem = pd.read_csv('cliente_baseline_memory_usage.txt', delim_whitespace=True, skiprows=3, usecols=[1, 3, 4], names=['kbmemfree', 'kbmemused', '%memused'])
cliente_backup_mem = pd.read_csv('cliente_backup_memory_usage.txt', delim_whitespace=True, skiprows=3, usecols=[1, 3, 4], names=['kbmemfree', 'kbmemused', '%memused'])

server_baseline_mem = pd.read_csv('server_baseline_memory_usage.txt', delim_whitespace=True, skiprows=3, usecols=[1, 3, 4], names=['kbmemfree', 'kbmemused', '%memused'])
server_backup_mem = pd.read_csv('server_backup_memory_usage.txt', delim_whitespace=True, skiprows=3, usecols=[1, 3, 4], names=['kbmemfree', 'kbmemused', '%memused'])


# Calcular los promedios de uso de CPU y memoria
cliente_baseline_cpu_avg = cliente_baseline_cpu.mean()
cliente_backup_cpu_avg = cliente_backup_cpu.mean()

server_baseline_cpu_avg = server_baseline_cpu.mean()
server_backup_cpu_avg = server_backup_cpu.mean()

cliente_baseline_mem_avg = cliente_baseline_mem.mean()
cliente_backup_mem_avg = cliente_backup_mem.mean()

server_baseline_mem_avg = server_baseline_mem.mean()
server_backup_mem_avg = server_backup_mem.mean()

# Crear gráficos comparativos
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Uso de CPU en Cliente
axs[0, 0].bar(['Baseline', 'Backup'], [cliente_baseline_cpu_avg['%user'], cliente_backup_cpu_avg['%user']], color='mediumturquoise', label='%user')
axs[0, 0].bar(['Baseline', 'Backup'], [cliente_baseline_cpu_avg['%system'], cliente_backup_cpu_avg['%system']], bottom=[cliente_baseline_cpu_avg['%user'], cliente_backup_cpu_avg['%user']], color='teal', label='%system')
axs[0, 0].set_title('Uso de CPU en Cliente')
axs[0, 0].set_ylabel('Porcentaje de Uso de CPU')
axs[0, 0].legend()

# Uso de CPU en Servidor
axs[0, 1].bar(['Baseline', 'Backup'], [server_baseline_cpu_avg['%user'], server_backup_cpu_avg['%user']], color='mediumturquoise', label='%user')
axs[0, 1].bar(['Baseline', 'Backup'], [server_baseline_cpu_avg['%system'], server_backup_cpu_avg['%system']], bottom=[server_baseline_cpu_avg['%user'], server_backup_cpu_avg['%user']], color='teal', label='%system')
axs[0, 1].set_title('Uso de CPU en Servidor')
axs[0, 1].set_ylabel('Porcentaje de Uso de CPU')
axs[0, 1].legend()

# Uso de Memoria en Cliente
axs[1, 0].bar(['Baseline', 'Backup'], [cliente_baseline_mem_avg['%memused'], cliente_backup_mem_avg['%memused']], color='mediumturquoise')
axs[1, 0].set_title('Uso de Memoria en Cliente')
axs[1, 0].set_ylabel('Porcentaje de Uso de Memoria')

# Uso de Memoria en Servidor
axs[1, 1].bar(['Baseline', 'Backup'], [server_baseline_mem_avg['%memused'], server_backup_mem_avg['%memused']], color='mediumturquoise')
axs[1, 1].set_title('Uso de Memoria en Servidor')
axs[1, 1].set_ylabel('Porcentaje de Uso de Memoria')

plt.tight_layout()
plt.show()
```
